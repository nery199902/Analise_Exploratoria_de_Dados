```python
# Definindo o DataFrame
df = pd.read_csv('HospInfo.csv')
```

```python
# Conferindo de carregou corretamente
df.head() # ou print(df.shape)
```

```python
# Tipos de colunas
colunas_numericas = df.select_dtypes(include=['int64', 'float64']).columns
colunas_categoricas = df.select_dtypes(include=['object', 'category']).columns

print('Numericas:', colunas_numericas)
print('Categoricas:', colunas_categoricas)
```

```python
# Valores Ausentes
print('\nValores ausentes:')
print(df.isnull().sum())
```

```python
# Tratamento dos valores Ausentes
colunas_footnote = [col for col in df.columns if 'footnote' in col]
df.drop(columns=colunas_footnote, inplace=True)

# Tratar valores ausentes restantes
df['County Name'] = df['County Name'].fillna('Desconhecido')
df['Meets criteria for meaningful use of EHRs'] = df['Meets criteria for meaningful use of EHRs'].fillna('Nao informado')

# Conferindo as alteracoes
print(df.isnull().sum())
```

```python
# Padronizando a coluna State
df['State'] = df['State'].str.strip().str.upper() #Deixando tudo em maiusculo e removendo espacos
```

```python
# Conferindo valores unicos
df['State'].unique()
```

```python
# Padronizando tipo de hospital - Limpando texto
df['Hospital Type'] = df['Hospital Type'].str.strip().str.title()
```

```python
df['Hospital Type'].unique() # Mostrar todos os valores sem repeticao
```

```python
# Conferindo resultado
print('Estados únicos:', df['State'].unique())
```

Criando um Dataframe com os dados por estado:

```python
# Convertendo String para Numerica # Pandas ignora NaN
df['Hospital overall rating'] = pd.to_numeric(
    df['Hospital overall rating'],
    errors='coerce'
)
df['Hospital overall rating'].isnull().sum() # Conferindo
```

```python
# Numero de hospitais por estado
hospitais_por_estado = df.groupby('State')['Hospital Name'].count()
# Media das avaliacoes por estado
media_avaliacao = df.groupby('State')['Hospital overall rating'].mean()
# Distribuicao dos tipos de hospital
distribuicao_tipos = df.groupby(['State', 'Hospital Type']).size()
```

Criando DataFrame consolidado:

```python
df_agg = df.groupby('State').agg({
    'Hospital Name': 'count',
    'Hospital overall rating': 'mean'
}).rename(columns={
    'Hospital Name': 'Numero de Hospitais',
    'Hospital overall rating': 'Media Avaliacao'
})
print(df_agg)
```

Criando uma nova variavel categorica para classificar os hospitais em 3 niveis de qualdiade:

```python
# Defininindo criterios
def classificar_qualidade(nota):
  if nota <= 2:
      return 'Baixa'
  elif nota == 3:
      return 'Media'
  elif nota >= 4:
      return 'Alta'
  else:
      return 'Nao Avaliado'
df['Nivel de Qualidade'] = df['Hospital overall rating'].apply(classificar_qualidade)

# Conferindo resultado
df[['Hospital overall rating', 'Nivel de Qualidade']].head()
```
